```python
from tqdm import tqdm

def basis_unigrid(data, mu):
    numBasisFncs = len(mu)+1
    Thetas = np.zeros((len(data), numBasisFncs))
    for ii in tqdm(range(len(data))):
        for jj in range(numBasisFncs):
            if jj == 0:
                Thetas[ii,jj] = 1
            else:
                Thetas[ii,jj] = np.exp( -( (((data['lat'][ii]-mu[jj-1,0])**2)/(2*(s[0]**2))) + \
                                       (((data['long'][ii]-mu[jj-1,1])**2)/(2*(s[1]**2))) ) )
    return Thetas
```
